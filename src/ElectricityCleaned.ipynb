```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Electricity.csv")
```

```python
commodities = np.unique(df["Commodity - Transaction"].values)
commodities
```

### Consumption by food and tobacco

```python
food = df[df["Commodity - Transaction"]== "Electricity - Consumption by food and tobacco"]
```

```python
food.set_index("Country or Area")
df_food = food.pivot_table(values = ["Quantity"], index=['Country or Area'], columns=['Year'])
df_food = df_food.dropna(thresh = 27)
df_food
```

```python
transport = df[df["Commodity - Transaction"]== "Electricity - Consumption by transport"]
transport.set_index("Country or Area")
df_transport = transport.pivot_table(values = ["Quantity"], index=['Country or Area'], columns=['Year'])
df_transport = df_transport.dropna(thresh = 27)
df_transport
```
